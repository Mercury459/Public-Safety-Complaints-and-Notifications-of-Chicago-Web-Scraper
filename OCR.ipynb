```python
import easyocr
import pathlib
import warnings
warnings.filterwarnings("ignore") 
```

```python
rec = []
r = []
reader = easyocr.Reader(['en'],gpu=False,model_storage_directory='./model')
for i in range(16150, 22728):
    path = pathlib.Path('C:/Users/25280/Downloads/NYPD/image/%d.png' %i)
    if not path.exists():
        continue
    result = reader.readtext('C:/Users/25280/Downloads/NYPD/image/%d.png' %i, detail = 0)
    j = 0
    for x in result:
        if "201" in x or "202" in x:
            r.append([wb["Name"][i], x])
```

```python
rec.append(r)
```

```python
# Remove incorrect results
c = []
for i in range(len(rec)):
    t = []
    for j in range(len(rec[i])):
        if rec[i][j] == ['s'] or rec[i][j] == []:
            continue
        t.append(rec[i][j])
    c.append(t)
```
